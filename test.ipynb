```python
import snowflake.connector as sf

conn = sf.connect(
    connection_name="myconnection",
)
```

```python
cursor = conn.cursor()

cursor.execute("SELECT current_version()")
one_row = cursor.fetchone()
print(one_row[0])
```

```python
cursor.execute("USE DATABASE COINS")
cursor.execute("USE SCHEMA TOPCOINSBYEXCHANGEVOLUME")

query = """
SELECT TOP 20 *
FROM
    TOPCOINSBYEXCHANGEVOLUME
ORDER BY INPUTDATE DESC
"""

data = cursor.execute(query)
df = data.fetch_pandas_all()
df
```

```python
cursor.execute("USE DATABASE COINS")
cursor.execute("USE SCHEMA TOPCOINSBYEXCHANGEVOLUME")

query = """
SELECT TOP 20 *
FROM
    TOPCOINSBYEXCHANGEVOLUME
ORDER BY INPUTDATE DESC, RANK
"""

data = cursor.execute(query)
tuList = data.fetchall()
tuList
```

```python
cursor.execute("USE DATABASE COINS")
cursor.execute("USE SCHEMA TOPCOINSBYEXCHANGEVOLUME")

query = """
SELECT TOP 1 INPUTDATE
FROM
    TOPCOINSBYEXCHANGEVOLUME
ORDER BY INPUTDATE DESC, RANK
"""

cursor.execute(query)
one_row = cursor.fetchone()
ddd = one_row[0]
print("Date = ", ddd)



query = f"""
SELECT TOP 20 *
FROM
    TOPCOINSBYEXCHANGEVOLUME
WHERE INPUTDATE='{ddd}'
ORDER BY INPUTDATE DESC, RANK
"""

data = cursor.execute(query)
tuList = data.fetchall()
tuList
```
